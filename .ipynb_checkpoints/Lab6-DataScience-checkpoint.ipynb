```python
import re
import time
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
from pattern.en import sentiment
from googletrans import Translator
```

```python
consumer_key = "XXXX"
consumer_secret = "XXXX"
access_token = "XXXX"
access_token_secret = "XXXX"
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
text_query = '#CoronavirusGT'
count = 500
try:
    # Creation of query method using parameters
    tweets = tweepy.Cursor(api.search,q=text_query).items(count)
    # Pulling information from tweets iterable object
    tweets_list = [[tweet.created_at,tweet.text] for tweet in tweets]
 
    # Creation of dataframe from tweets list
    # Add or remove columns as you remove tweet information
    tweets_df = pd.DataFrame(tweets_list)
except BaseException as e:
    print('failed on_status,',str(e))
    time.sleep(3)
```

```python
tweets = tweets_df[1]
tweets= tweets.apply(lambda x: x.lower() if type(x) == str else x)
tweets = tweets.apply(lambda x: re.sub('[^a-zA-Z0-9 ]+','', x) if type(x) == str else x)
tweets = tweets.apply(lambda x: re.sub('\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*','',x) if type(x) == str else x)
```

### Exploración

Frecuencia de palabras

```python
# Las veinte palabras más populares sin conectores
frecuencia = Counter()
for i in tweets:
    frecuencia.update(w for w in str(i).split() if len(w) > 3)
    
frecuencia.most_common(20)
```

```python
# Nube de las veinte palabras más populares sin conectores
nube = WordCloud(background_color="white",width=4000,height=2000, max_words=20,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(frecuencia)
plt.imshow(nube)
```

Promedio de caracteres por tweet

```python
# Largo promedio de los tweets
lenght = 0
for i in tweets:
    lenght += len(i)

average = lenght/len(tweets)

print("Promedio de caracteres por tweet: " + str(average))
```

Promedio de palabras por tweet

```python
# Largo promedio de los tweets
lenght = 0
for i in tweets:
    lenght += len(i.split())

average = lenght/len(tweets)

print("Promedio de palabras por tweet: " + str(average))
```

N-grams

```python
def calc_3gram(data):
    
    n_gram = []
    
    for tmp_data in data:
        tmp_data = str(tmp_data)
        last = ""
        p_last = ""
        for wn in tmp_data.split():
            if p_last == "":
                p_last = wn
                continue
            elif last == "":
                last = p_last
                p_last = wn
                continue
                
            n_gram.append((last, p_last, wn))
            last = p_last
            p_last = wn

    return n_gram
```

```python
# Trigrama
n3_gram = calc_3gram(tweets)
```

Sentimiento en palabras

```python
# Se utilizara un traductor para traducir las palabras a ingles
translator = Translator()
```

```python
# positividad de los tweets
tweets_positivos = 0
tweets_negativos = 0
tweets_neutrales = 0

for i in tweets:
    positividad = 0
    
    i_eng = translator.translate(i).text
    
    for w in i_eng.split():
        s = sentiment(w)[0]
        positividad += s
        
    if positividad > 0:
        tweets_positivos += 1
    elif positividad < 0:
        tweets_negativos += 1
    else:
        tweets_neutrales += 1

print("Tweets positivos: " + str(tweets_positivos))
print("Tweets negativos: " + str(tweets_negativos))
print("Tweets neutrales: " + str(tweets_neutrales))
print("Total de tweets: " + str(len(tweets)))
```
